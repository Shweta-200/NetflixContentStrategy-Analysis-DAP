**categorizing Vote_Average column**

```python
def catigorize_col(df, col, labels):
    """
    Categorizes a specified column in a DataFrame based on quartiles.

    Args:
        df (pd.DataFrame): The DataFrame to modify.
        col (str): The name of the column to categorize.
        labels (list): The labels to use for the categories.

    Returns:
        pd.DataFrame: The modified DataFrame.
    """
    # setting the edges to cut the column accordingly
    edges = [
        df[col].describe()['min'],
        df[col].describe()['25%'],
        df[col].describe()['50%'],
        df[col].describe()['75%'],
        df[col].describe()['max']
    ]
    df[col] = pd.cut(df[col], edges, labels=labels, duplicates='drop')
    return df
```

```python
# Convert 'Vote_Average' to numerical representation before calculating correlation.
# Assuming 'Vote_Average' categories have an order:
# 'not_popular', 'below_avg', 'average', 'popular'
vote_avg_mapping = {
    'not_popular': 1,
    'below_avg': 2,
    'average': 3,
    'popular': 4
}
df['Vote_Average_Num'] = df['Vote_Average'].map(vote_avg_mapping)

# Now use the numerical representation for correlation
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Convert 'Release_Date' to datetime format
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')

# Extract year from 'Release_Date'
df['Release_Year'] = df['Release_Date'].dt.year

# Count movies released per year
yearly_counts = df['Release_Year'].value_counts().sort_index()

# Plot the number of movies released over time
plt.figure(figsize=(10, 5))
sns.lineplot(x=yearly_counts.index, y=yearly_counts.values, marker="o", linestyle="-", color="b")
plt.title('Number of Movies Released Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count movies per genre
genre_counts = df.explode('Genre')['Genre'].value_counts()

# Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Movies by Genre')
plt.axis('equal')  # Ensures a circular pie chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count movies per language (show top 10 for clarity)
top_languages = df['Original_Language'].value_counts().head(10)

# Plot pie chart
plt.figure(figsize=(10, 7))  # Increased size
wedges, texts, autotexts = plt.pie(
    top_languages,
    labels=top_languages.index,
    autopct='%1.1f%%',
    startangle=140,
    wedgeprops={'edgecolor': 'black'},  # Add edge color for clarity
    textprops={'fontsize': 12}  # Increase font size
)

# Adjust label text sizes
for text in texts + autotexts:
    text.set_fontsize(10)  # Increase font size

# Add a legend outside the chart for better visibility
plt.legend(top_languages.index, title="Languages", bbox_to_anchor=(1.05, 1), loc="upper left")

# Title and display
plt.title('Top 10 Languages in Movies')
plt.axis('equal')  # Keeps it circular
plt.show()
```
